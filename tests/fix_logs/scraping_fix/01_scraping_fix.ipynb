```python
pricefix_df = pd.read_csv("null_prices_fix.csv")
pricefix_df.head()
```

```python
#Función para convertir a decimales los objets, tomando los primeros cinco caracteres
def clean_and_convert(value):
    if isinstance(value, float):  
        return value
    return pd.to_numeric(str(value)[:5].replace('%', ''), errors='coerce')

# Aplicar la conversión a las columnas deseadas
columns_to_convert = ['body', 'tannis', 'sweetness', 'acidity']
for col in columns_to_convert:
    wines_df[col] = wines_df[col].apply(clean_and_convert)
    pricefix_df[col] = pricefix_df[col].apply(clean_and_convert)

wines_df.info()
```

```python
pricefix_df.info()
```

```python
# Fixed Price Quality Check (Tastes)
fixed_df = wines_df.join(pricefix_df.set_index("wine_link"), on="wine_link", rsuffix="_fix")

col1 = "sweetness"
col2 = col1 + "_fix"

fix_df = fixed_df[fixed_df["price"].isna() & (fixed_df[col1] != fixed_df[col2])][[col1, col2]]

fix_df["dif"] = np.where(fix_df[col1].isna() | fix_df[col2].isna(), 1, (fix_df[col2]/fix_df[col1]) - 1)

fix_df[(fix_df["dif"]>0.01) & (~fix_df[col1].isna() | ~fix_df[col1].isna())]
```

```python
# Fixed Price Quality Check (Pairings)
# Existen algunos vinos a los cuales en la segunda búsqueda no les aparecen los pairings
fixed_df = wines_df.join(pricefix_df.set_index("wine_link"), on="wine_link", rsuffix="_fix")

col1 = "pairings"
col2 = col1 + "_fix"

fix_df = fixed_df[fixed_df["price"].isna() & (fixed_df[col1] != fixed_df[col2])][["wine_link", col1, col2]]

fix_df.reset_index(drop=True)
```

```python
# Fixed Price Quality Check (Grapes)
# Existen algunos que tienen las grapes mal scrapeadas
fixed_df = wines_df.join(pricefix_df.set_index("wine_link"), on="wine_link", rsuffix="_fix")

col1 = "grapes"
col2 = col1 + "_fix"

fix_df = fixed_df[fixed_df["price"].isna() & (fixed_df[col1] != fixed_df[col2])][["wine_link", col1, col2]]

fix_df
```

```python
# Fixed Price Quality Check (Region)
# Existen algunos que tienen las Regions mal scrapeadas
fixed_df = wines_df.join(pricefix_df.set_index("wine_link"), on="wine_link", rsuffix="_fix")

col1 = "region"
col2 = col1 + "_fix"

fix_df = fixed_df[fixed_df["price"].isna() & (fixed_df[col1] != fixed_df[col2])][["wine_link", col1, col2]]

fix_df
```
